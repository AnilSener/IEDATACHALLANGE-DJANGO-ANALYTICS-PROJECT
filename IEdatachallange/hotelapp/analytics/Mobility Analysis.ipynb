```python
import pandas as pd
import numpy as np
from bson import ObjectId
from pymongo import MongoClient
from planar import BoundingBox
```

```python
client = MongoClient('localhost',27017)
db = client.hotelapp
mobile = db.map_item
attractions = db.attraction
```

```python
loc = pd.DataFrame(list(mobile.find({"properties.hour":0, "properties.date":"2014-01-01"}, {"properties.hour":1, "properties.n_people":1, "properties.date":1,
                                         "geometry":1})))
att = pd.DataFrame(list(attractions.find({"_id":{"$exists": "true"}})))
```

```python
loc.head()
```

```python
attcor = []
for i in range(len(att['geometry'].tolist())):
    try:
        attcor.append([att['geometry'].values.tolist()[i]['coordinates'],att['properties'][i]['locationName']])
    except Exception: #some are nan so need to skip
        continue
attcor = pd.DataFrame(attcor, columns=['coor', 'name'])
```

```python
attIn = []
cors = attcor.ix[:,0]
names = attcor.ix[:,1]
for x in range(loc.shape[0]):
    a = loc.ix[x,1]['coordinates'][0][0]
    b = loc.ix[x,1]['coordinates'][0][1]
    c = loc.ix[x,1]['coordinates'][0][2]
    d = loc.ix[x,1]['coordinates'][0][3]
    e = loc.ix[x,1]['coordinates'][0][4]
    bbox= BoundingBox([a,b,c,d,e])
    for i in range(len(cors)):
        if bbox.contains_point(cors[i]):
            if loc.ix[x,0] in attIn:
                pass
            else:
                attIn.append([loc.ix[x,0], names[i]])
```

```python
data= []
for i in range(gug.shape[0]):
    data.append([gug.ix[i,2]['date'], gug.ix[i,2]['hour'], gug.ix[i,2]['n_people']])
data = pd.DataFrame(data, columns= ['date', 'hour', 'n_people'])
from datetime import time
for i in range(data.shape[0]):
   data['hour'][i] = str(time(data['hour'][i]))
data['date_time'] = pd.to_datetime(data['date'] + ' ' + data['hour'])
data = data.sort('date_time')
```

```python
import matplotlib.pyplot as plt
from matplotlib import dates
import datetime as dt
import numpy as np

plt.plot(data.ix['2014-01-22':'2014-01-23',].index, data.ix['2014-01-22':'2014-01-23',2])

plt.rcParams["figure.figsize"] = (12,6)

xticks = dates.date2num(data['date_time'].tolist())
#plt.xticks(xticks)
plt.grid(True)
```

```python
import statsmodels.api as sm
from statsmodels.graphics.api import qqplot
from statsmodels.tsa.arima_model import ARMA
```

```python
data.index = data['date_time'].values
armaMod = sm.tsa.ARMA(data.ix['2014-01-01':'2014-01-29',2], (5,0,1)).fit()
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111)
ax = armaMod.resid.plot(ax=ax)
```

```python
resid = armaMod.resid
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(resid.values.squeeze(), lags=40, ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(resid, lags=40, ax=ax2)
```

```python
yhat = armaMod.predict('2014-01-29', '2014-01-31')
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(data.ix['2014-01-29':,2], 'r-' )
ax.plot(yhat, 'b-')
```
