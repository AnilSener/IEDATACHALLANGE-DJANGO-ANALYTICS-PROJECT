```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
import math
import networkx as nx
from networkx.algorithms import bipartite
from matplotlib import pyplot as plt
%matplotlib inline
def getHotelName(dictObj):
    return dictObj.get("name")
def getAttractionName(dictObj):
    return dictObj.get("locationName")
```

```python
client = MongoClient('localhost',27017)
db = client.hotelapp
box=[[-3.4492757,42.9818774], [-2.4128145,43.4568551]]
dfTAAttractions = pd.DataFrame(list(db.attraction.find({"geometry.coordinates":{"$geoWithin":{"$box":box}}})))
dfTAHotels = pd.DataFrame(list(db.t_a_hotel.find({"geometry.coordinates":{"$geoWithin":{"$box":box}}})))
dfTAUsers = pd.DataFrame(list(db.t_a_user.find({"_id":{"$exists":"true"}})))
dfTAHotelReviews = pd.DataFrame(list(db.hotel_review.find({"_id":{"$exists":"true"}})))
dfTAAttractionReviews = pd.DataFrame(list(db.review.find({"_id":{"$exists":"true"}})))
dfHotels = pd.DataFrame(list(db.hotel.find({"geometry":{"$geoWithin":{"$box":box}}})))
dfHotels["hotelID"]=dfHotels["hotelID"].astype('int').astype('str')
dfTAAttractions["attractionID"]=dfTAAttractions["attractionID"].astype('str')
```

```python
print(len(dfTAHotels))
dfTAHotels["hotelID"]=float("NaN")
validlist=[]
for i,tahotel in enumerate(dfTAHotels["hotelName"].values):
    hotelID=-1
    for j,hotel in enumerate(dfHotels["properties"].values):
        lat=dfHotels.loc[j]["geometry"].get("coordinates")[1]
        lon=dfHotels.loc[j]["geometry"].get("coordinates")[0]
        if hotel.get("name") in tahotel:
            
            hotelID=dfHotels.loc[j]["hotelID"]
            break;
        elif (lat==dfTAHotels.loc[i]["geometry"].get("coordinates")[1] and lon==dfTAHotels.loc[i]["geometry"].get("coordinates")[0]):
            
            hotelID=dfHotels.loc[j]["hotelID"]
            break;
    if int(hotelID)>-1:    
        dfTAHotels["hotelID"].loc[[i]]=str(hotelID)
        
#print(len(dfTAHotels))
dfTAHotels.dropna(inplace=True)
#dfTAHotelReviews=dfTAHotelReviews[dfTAHotelReviews["reviewLocation"].isin(dfTAHotels["_id"].values)]
#print(len(dfTAHotels))
```

```python
#dfTAHotelUsers=dfTAUsers[dfTAUsers["_id"].isin(dfTAHotelReviews["user"].values)]
#print(len(dfTAUsers))
#print(len(dfTAHotelUsers))
#dfTAAttractionUsers=dfTAUsers[dfTAUsers["_id"].isin(dfTAAttractionReviews["user"].values)]
#print(len(dfTAUsers))
#print(len(dfTAAttractionUsers))
#dfTAReducedUsers=dfTAHotelUsers[dfTAHotelUsers["_id"].isin(dfTAAttractionReviews["user"].values)]
#print(len(dfTAUsers))
#print(len(dfTAReducedUsers))

dfUserHReviewEdge=pd.merge(dfTAUsers,dfTAHotelReviews,how='inner',left_on="_id",right_on="user")[["name","_id_y"]]
dfUserHReviewEdge.rename(columns={'_id_y': 'reviewID'}, inplace=True)
dfHotelHReviewEdge=pd.merge(dfTAHotels,dfTAHotelReviews,how='inner',left_on="_id",right_on="reviewLocation")[["hotelID","_id_y"]]
dfHotelHReviewEdge.rename(columns={'_id_y': 'reviewID'}, inplace=True)
#hotel-user merge
dfHotelUserRaw=pd.merge(dfHotelHReviewEdge,dfUserHReviewEdge,how='inner',left_on="reviewID",right_on="reviewID")[["hotelID","name","reviewID"]]
dfHotelUserEdge=dfHotelUserRaw.groupby(["hotelID","name"]).count().reset_index()
dfHotelUserEdge.rename(columns={'reviewID':"reviewCount"}, inplace=True)

dfUserAReviewEdge=pd.merge(dfTAUsers,dfTAAttractionReviews,how='inner',left_on="_id",right_on="user")[["name","_id_y"]]
dfUserAReviewEdge.rename(columns={'_id_y': 'reviewID'}, inplace=True)
dfAttractionAReviewEdge=pd.merge(dfTAAttractions,dfTAAttractionReviews,how='inner',left_on="_id",right_on="reviewLocation")[["attractionID","_id_y"]]
dfAttractionAReviewEdge.rename(columns={'_id_y': 'reviewID'}, inplace=True)
#Attraction-user merge
dfAttractionUserRaw=pd.merge(dfAttractionAReviewEdge,dfUserAReviewEdge,how='inner',left_on="reviewID",right_on="reviewID")[["attractionID","name","reviewID"]]
dfAttractionUserEdge=dfAttractionUserRaw.groupby(["attractionID","name"]).count().reset_index()
dfAttractionUserEdge.rename(columns={'reviewID':"reviewCount"}, inplace=True)
#print(len(dfUserHReviewEdge))

#Hotel Attraction Merge
dfHotelAttractionRaw=pd.merge(dfHotelUserEdge,dfAttractionUserEdge,how='inner',on="name")[["hotelID","attractionID","reviewCount_y"]]
dfHotelAttractionEdge=dfHotelAttractionRaw.groupby(["hotelID","attractionID"]).count().reset_index()
dfHotelAttractionEdge.rename(columns={"reviewCount_y":"userCount"}, inplace=True)
```

HOTEL - USER BIPARTITE GRAPH

```python
g = nx.Graph()

hotel_user_edges=[(e[0],e[1],e[2]) for e in dfHotelUserEdge.values]
g.add_nodes_from(np.unique(dfHotelUserEdge["name"].values), bipartite=0,color="blue") # Add the node attribute "bipartite"
g.add_nodes_from(np.unique(dfHotelUserEdge["hotelID"].values), bipartite=1,color="green")
g.add_weighted_edges_from(hotel_user_edges)
#nx.draw_networkx(g,node_color=["blue","green"],with_labels=False,limits=plt.axis('off'))
#nx.draw_circular(g,node_color=["blue","green"],with_labels=False)
degree = bipartite.degree_centrality(g,bipartite.sets(g)[0])
fig = plt.gcf()
fig.set_size_inches(10,10)

nx.draw_networkx(g,limits=plt.axis('off'),node_color=["blue","green"],node_size = [v * 1500 for v in degree.values()],with_labels=False)
```

```python
hotelusernet=bipartite.weighted_projected_graph(g, np.unique(dfHotelUserEdge["name"].values), ratio=False)

#hotelusernet=list(nx.connected_component_subgraphs(hotelusernet))[0]

weights=[math.log(edata['weight']) for f,t,edata in hotelusernet.edges(data=True)]
#Finally, we draw the graph:


#node_size=[e*1000+10 for e in eigenvector_centrality.values()]
fig = plt.gcf()
fig.set_size_inches(10,10)
nx.draw_networkx(hotelusernet, edge_color=weights,limits=plt.axis('off'),with_labels=False)

#print(['%s %0.2f'%(node,centrality[node]) for node in centrality])
```

Updating Trip Advisor Users Table with centrality calculations needed to rank influential users in terms of hotel reviews.

```python
eigenvector_centrality=nx.eigenvector_centrality(hotelusernet)
degree_centrality=nx.degree_centrality(hotelusernet)
betweenness_centrality=nx.betweenness_centrality(hotelusernet)
for node in eigenvector_centrality:
    db.t_a_user.update({'name':node},{"$set":{'hotel_eigenvector_centrality':eigenvector_centrality[node],'hotel_degree_centrality':degree_centrality[node],'hotel_betweenness_centrality':betweenness_centrality[node]}},upsert=False)
```

```python
hotelnet=bipartite.weighted_projected_graph(g, np.unique(dfHotelUserEdge["hotelID"].values), ratio=False)

pos=nx.spring_layout(hotelnet) # positions for all nodes
#hotelnet=list(nx.connected_component_subgraphs(hotelnet))[0]
weights=[edata['weight'] for f,t,edata in hotelnet.edges(data=True)]
hotelnames=list(map(getHotelName,dfHotels[dfHotels["hotelID"].isin(hotelnet.nodes())]["properties"].values))
hoteldict=dict(zip(list(hotelnet.nodes()), hotelnames))
#nx.draw_networkx(hotelnet,pos=pos,font_size=8, edge_color=weights,limits=plt.axis('off'))
eigenvector_centrality=nx.eigenvector_centrality(hotelnet)
degree_centrality=nx.degree_centrality(hotelnet)
betweenness_centrality=nx.betweenness_centrality(hotelnet)
node_size=[e*1000+10 for e in list(degree_centrality.values())]
fig = plt.gcf()
fig.set_size_inches(10,10)

nodes=nx.draw_networkx_nodes(hotelnet, pos=pos, nodelist = hotelnet.nodes(),node_size=node_size)
edges=nx.draw_networkx_edges(hotelnet, pos=pos, edgelist = hotelnet.edges(),width=weights,edge_color=weights)
nx.draw_networkx_labels(hotelnet, pos=pos,labels=hoteldict,font_size=10)
nx.draw_spectral(hotelnet, with_labels = True)
plt.sci(edges)
plt.colorbar()
```

ATTRACTION -USER NETWORK

```python
g2 = nx.Graph()

attraction_user_edges=[(e[0],e[1],e[2]) for e in dfAttractionUserEdge.values]
g2.add_nodes_from(np.unique(dfAttractionUserEdge["name"].values), bipartite=0) # Add the node attribute "bipartite"
g2.add_nodes_from(np.unique(dfAttractionUserEdge["attractionID"].values), bipartite=1)
g2.add_weighted_edges_from(attraction_user_edges)
nx.draw_networkx(g2,node_color=["blue","green"],with_labels=False,limits=plt.axis('off'))
```

```python
attractionusernet=bipartite.weighted_projected_graph(g2, np.unique(dfAttractionUserEdge["name"].values), ratio=False)

weights=[math.log(edata['weight']) for f,t,edata in attractionusernet.edges(data=True)]
#Finally, we draw the graph:
eigenvector_centrality=nx.eigenvector_centrality(attractionusernet)
degree_centrality=nx.degree_centrality(attractionusernet)
betweenness_centrality=nx.betweenness_centrality(attractionusernet)
nx.draw_networkx(attractionusernet, edge_color=weights,limits=plt.axis('off'))
```

```python
for node in eigenvector_centrality:
    db.t_a_user.update({'name':node},{"$set":{'attraction_eigenvector_centrality':eigenvector_centrality[node],'attraction_degree_centrality':degree_centrality[node],'attraction_betweenness_centrality':betweenness_centrality[node]}},upsert=False)
```

```python
attractionnet=bipartite.weighted_projected_graph(g2, np.unique(dfAttractionUserEdge["attractionID"].values), ratio=False)
attractionnet=list(nx.connected_component_subgraphs(attractionnet))[1]
pos=nx.spring_layout(attractionnet) # positions for all nodes
weights=[edata['weight'] for f,t,edata in attractionnet.edges(data=True)]
attractionnames=list(map(getAttractionName,dfTAAttractions[dfTAAttractions["attractionID"].isin(attractionnet.nodes())]["properties"].values))
attractiondict=dict(zip(attractionnet.nodes(), attractionnames))
fig = plt.gcf()
fig.set_size_inches(10,10)
nodes=nx.draw_networkx_nodes(attractionnet, pos=pos, nodelist = attractionnet.nodes())
edges=nx.draw_networkx_edges(attractionnet, pos=pos, edgelist = attractionnet.edges(),width=weights,edge_color=weights)
nx.draw_networkx_labels(attractionnet, pos=pos,labels=attractiondict,font_size=8)
nx.draw_spectral(attractionnet, with_labels = True)
plt.sci(edges)
plt.colorbar()
#nx.draw_networkx_labels(attractionnet,pos=pos,labels=attractiondict,font_size=8, edge_color=weights,limits=plt.axis('off'))
```
